# 受けた力を出力

```python
import hebi
from time import sleep
import math
from math import pi
import scipy.constants
import numpy as np
import threading
import scipy.constants

lookup = hebi.Lookup()
group = lookup.get_group_from_names(['Test'], ['Test'])
group_feedback = hebi.GroupFeedback(group.size)
group_command = hebi.GroupCommand(group.size)
group_command.reference_effort = 0.0000

if group is None:
    print('Group not found: Did you forget to set the module family and names above?')
    exit(1)

for entry in lookup.entrylist:
    print(entry)

            
class MyThread(threading.Thread):
    
    def __init__(self):
        
        super(MyThread, self).__init__()
        self.stop_event = threading.Event()
        self.setDaemon(True)
        
        
    def stop(self):
        
        self.stop_event.set()

        
    def run(self):
        
        group.feedback_frequency = 100.0

        while not self.stop_event.is_set():
            
            group.get_next_feedback(reuse_fbk=group_feedback)
            effort = group_feedback.effort

            group_command.effort = -effort
            print('Effort : {0}'.format(group_command.effort))
            group.send_command(group_command)
            
        print('END')
        group_command.effort = 0.0
        group.send_command(group_command)
        group.clear_feedback_handlers()
        
    
if __name__ == '__main__':
    th = MyThread()
    th.start()
    sleep(10)
    th.stop()
    sleep(1)
    group.clear_feedback_handlers()
```

```python
def get_grav_comp_efforts(robot_model, positions, gravityVec): # gravityVec = [0.0, 0.0, 1.0]
    # Normalize gravity vector (to 1g, or 9.8 m/s^2)
    normed_gravity = gravityVec / numpy.linalg.norm(gravityVec) * 9.81

    jacobians = robot_model.get_jacobians('CoM', positions)
    # Get torque for each module
    # comp_torque = J' * wrench_vector
    # (for each frame, sum this quantity)
    comp_torque = numpy.zeros((robot_model.dof_count, 1))

    # Wrench vector
    wrench_vec = numpy.zeros(6)  # For a single frame; this is (Fx/y/z, tau x/y/z)
    num_frames = robot_model.get_frame_count('CoM')

    for i in range(num_frames):
        # Add the torques for each joint to support the mass at this frame
        wrench_vec[0:3] = normed_gravity * robot_model.masses[i]
        comp_torque += jacobians[i].transpose() * numpy.reshape(wrench_vec, (6, 1))

    return numpy.squeeze(comp_torque)
```

# 押したら動き出す

```python
import hebi
from time import sleep
import math
from math import pi
import scipy.constants
import numpy as np
import threading
import scipy.constants

lookup = hebi.Lookup()
group = lookup.get_group_from_names(['Test'], ['Test'])
group_feedback = hebi.GroupFeedback(group.size)
group_command = hebi.GroupCommand(group.size)
group_command.reference_effort = 0.0000

if group is None:
    print('Group not found: Did you forget to set the module family and names above?')
    exit(1)

for entry in lookup.entrylist:
    print(entry)

            
class MyThread(threading.Thread):
    
    def __init__(self):
        
        super(MyThread, self).__init__()
        self.stop_event = threading.Event()
        self.setDaemon(True)
        
        
    def stop(self):
        
        self.stop_event.set()

        
    def run(self):
        
        group.feedback_frequency = 100.0
        init_eff = group.get_next_feedback().effort
        sw = False

        while not self.stop_event.is_set():
            
            group.get_next_feedback(reuse_fbk=group_feedback)
            eff = group_feedback.effort
            dif = abs(eff[0] - init_eff[0])
            
            if sw == False:
                if dif >= 0.2:
                    sw = True
            else:
                group_command.velocity = 1 # rad/sec
                group.send_command(group_command)
            
        print('END')
        group_command.velocity = 0.0
        group.send_command(group_command)
        
    
if __name__ == '__main__':
    th = MyThread()
    th.start()
    sleep(10)
    th.stop()
    sleep(1)
```

# 放置したら減速する

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

#================================================
#   歩行器をキーボード入力で回転速度を可変にした
#================================================

import hebi
import math
from math import pi
import numpy as np
import threading
from getch import getch
from time import sleep, time
import csv

lookup = hebi.Lookup()
group = lookup.get_group_from_names(['Wheel'], ['Right'])

if group is None:
    print('Group not found: Did you forget to set the module family and names above?')
    exit(1)

for entry in lookup.entrylist:
    print(entry)
            
class MyThread(threading.Thread):
    
    def __init__(self):
        
        super(MyThread, self).__init__()
        self.com = np.zeros(3)
        self.stop_event = threading.Event()
        self.setDaemon(True)
        
        
    def stop(self):
        
        self.stop_event.set()

        
    def run(self):
        
        group_feedback = hebi.GroupFeedback(group.size)
        group_command = hebi.GroupCommand(group.size)

        group.feedback_frequency = 10
        group_command.reference_effort = 0.0000
        
        new_velocity_kp_gains = [0.004]

        for new_gain in new_velocity_kp_gains:
            for i in range(group.size):
                group_command.velocity_kp = new_gain
            if not group.send_command_with_acknowledgement(group_command):
                print('Did not get acknowledgement from module when sending gains. Check connection.')
                exit(1)
            print('Set velocity kP gain to: {0}'.format(new_gain))
        
        print("=========================================\n  q  w  e  \n  a  s  d  \n     x     \n=========================================\n\nIf you want to quit, please press 'Ctrl + c'.\n")
        
        start = time()
        g = 9.8 # 重力
        u = 0.4 #摩擦係数
        start_time = start
        init_eff = group.get_next_feedback().effort
        
        
        while not self.stop_event.is_set():
            
            group.get_next_feedback(reuse_fbk=group_feedback)
            x = self.com[0]
            yL = self.com[1]
            yR = self.com[2]

            spring_constant = x # rad/sec
            group_command.velocity = [-spring_constant * yR]
            group.send_command(group_command)
            
            speed = group_command.velocity[0]
            t = time() - start
            if t >= 0.01:

                speed = speed ** 2 - 2*u*g*t # 速度の2乗

                if speed > 0:
                    speed = speed ** 0.5
                else:
                    speed = 0.0

                print("speed: {0}".format(speed))
                start = time()
            
        
    def Move(self, key):
        
        key = key
        x = 0.1
        y = 0.5
        com = self.com
        
        if key == 119: # w 前進
            com[0] += x
            com[1] = 1.0
            com[2] = 1.0
        elif key == 113: # q 左折
            com[1] = y
            com[2] = 1.0
        elif key == 101: # e 右折
            com[1] = 1.0
            com[2] = y
        elif key ==115: # s 停止
            while not com[0]==0:
                if com[0] >= 0.1:
                    com[0] -= x
                elif com[0] <= -0.1:
                    com[0] += x
                else:
                    com[0] = 0.0
            com[1] = 0.0
            com[2] = 0.0
        elif key == 97: # a 左旋回
            if com[0] >= 0.0:
                com[1] = -1.0
                com[2] = 1.0
            elif com[0] < -0.0:
                com[1] = 1.0
                com[2] = -1.0
        elif key == 100: # d 右旋回
            if com[0] >= 0.0:
                com[1] = 1.0
                com[2] = -1.0
            elif com[0] < -0.0:
                com[1] = -1.0
                com[2] = 1.0
        elif key == 120: # x 後退
            com[0] -= x
            com[1] = 1.0
            com[2] = 1.0
        
        if com[0] > pi:
            com[0] = 3.0
        elif com[0] < -pi:
            com[0] = -3.0
            
        
        print('x = {0:.2f}, yL = {1:.2f}, yR = {2:.2f}'.format(com[0], com[1], com[2]))
        
if __name__ == '__main__':
    th = MyThread()
    th.start()
    
    while True:
        key = ord(getch())
        if key == 113: # q
            th.stop()
            sleep(1)
            break
        else:
            th.Move(key)
            
    print('END')
    group_command = hebi.GroupCommand(group.size)
    group_command.velocity = [0.0]
    group.send_command(group_command)
```
